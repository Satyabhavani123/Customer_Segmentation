# Customer Segmentation with RFM + K-Means 


## 1) Setup
- If running locally, install dependencies with `pip install -r requirements.txt`.


```python
# --- Imports & Versions ---
import sys, os
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

print("Python:", sys.version)
print("Pandas:", pd.__version__)
print("NumPy:", np.__version__)

# Plotting defaults (Matplotlib only, no styles/colors)
plt.rcParams["figure.figsize"] = (8, 5)
```

## 2) Load Data


```python
# --- Load Data ---
USE_SAMPLE_DATA = True  # Set to False to load your own CSV

if USE_SAMPLE_DATA:
    csv_path ="C:\\Users\\BHAVANI\\Downloads\\customer-segmentation-rfm\\customer-segmentation-rfm\\sample_transactions.csv"
else:
    # Upload or set path to your own file, e.g.:
    # csv_path = "/path/to/your/ecommerce_data.csv"
    csv_path = "C:\\Users\\BHAVANI\\Downloads\\customer-segmentation-rfm\\customer-segmentation-rfm\\sample_transactions.csv"  # placeholder

df = pd.read_csv(csv_path, parse_dates=["InvoiceDate"])
print("Shape:", df.shape)
df.head()
```

## 3) Clean & Prepare
- Drop missing `CustomerID`
- Remove returns/negative quantities and zero/negative prices
- Ensure `InvoiceDate` is datetime
- Compute `TotalAmount = Quantity * UnitPrice`


```python
# --- Cleaning ---
before = df.shape[0]
df = df.dropna(subset=["CustomerID"])
df = df[(df["Quantity"] > 0) & (df["UnitPrice"] > 0)]
df["TotalAmount"] = df["Quantity"] * df["UnitPrice"]
after = df.shape[0]

print(f"Rows before: {before}, after clean: {after}")
df[["InvoiceNo","CustomerID","InvoiceDate","Quantity","UnitPrice","TotalAmount"]].head()
```

## 4) RFM Feature Engineering
- **Recency (R):** Days since last purchase (lower is better)
- **Frequency (F):** Number of invoices
- **Monetary (M):** Total spending


```python
# --- RFM Calculation ---
snapshot_date = df["InvoiceDate"].max() + pd.Timedelta(days=1)

rfm = df.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (snapshot_date - x.max()).days,
    "InvoiceNo": "nunique",
    "TotalAmount": "sum"
}).rename(columns={
    "InvoiceDate": "Recency",
    "InvoiceNo": "Frequency",
    "TotalAmount": "Monetary"
}).reset_index()

print("RFM shape:", rfm.shape)
rfm.head()
```

## 5) Scale RFM (important for K-Means)


```python
# --- Scaling ---
features = ["Recency","Frequency","Monetary"]
X = rfm[features].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Scaled sample (first 5 rows):")
X_scaled[:5]
```

## 6) Choose K (number of clusters)


```python
# --- Elbow & Silhouette ---
sse = []
sil = []
K_vals = list(range(2, 9))

for k in K_vals:
    km = KMeans(n_clusters=k, random_state=42, n_init="auto")
    km.fit(X_scaled)
    sse.append(km.inertia_)
    labels = km.labels_
    sil.append(silhouette_score(X_scaled, labels))

# Elbow plot
plt.figure()
plt.plot(K_vals, sse, marker="o")
plt.title("Elbow Method (SSE vs K)")
plt.xlabel("K")
plt.ylabel("SSE")
plt.show()

# Silhouette plot
plt.figure()
plt.plot(K_vals, sil, marker="o")
plt.title("Silhouette Score vs K")
plt.xlabel("K")
plt.ylabel("Silhouette Score")
plt.show()

best_k = K_vals[int(np.argmax(sil))]
print("Best K by Silhouette (guide):", best_k)
```

## 7) Fit Final K-Means and Create Segments



```python
# --- Final K-Means ---
K = best_k  # You may override, e.g., K = 5
kmeans = KMeans(n_clusters=K, random_state=42, n_init="auto")
rfm["Cluster"] = kmeans.fit_predict(X_scaled)

# Basic profiling
profile = rfm.groupby("Cluster")[["Recency","Frequency","Monetary"]].mean().round(2)
counts = rfm["Cluster"].value_counts().sort_index()
print("Cluster sizes:\n", counts)
print("\nCluster means:\n", profile)

rfm.head()
```

## 8) Quick Visuals (Matplotlib)


```python
# --- Scatter plots ---
plt.figure()
plt.scatter(rfm["Frequency"], rfm["Monetary"], c=rfm["Cluster"])
plt.xlabel("Frequency")
plt.ylabel("Monetary")
plt.title("Clusters: Frequency vs Monetary")
plt.show()

plt.figure()
plt.scatter(rfm["Recency"], rfm["Monetary"], c=rfm["Cluster"])
plt.xlabel("Recency (days)")
plt.ylabel("Monetary")
plt.title("Clusters: Recency vs Monetary")
plt.show()
```

## 9) (Optional) Name Segments for Business


```python
# --- Optional: Name segments ---
# We'll name based on relative cluster means
prof = rfm.groupby("Cluster")[["Recency","Frequency","Monetary"]].mean()

# Rank clusters for each metric (lower Recency is better)
rec_rank = prof["Recency"].rank(ascending=True)
freq_rank = prof["Frequency"].rank(ascending=False)
mon_rank = prof["Monetary"].rank(ascending=False)

score = rec_rank + freq_rank + mon_rank  # lower is better
order = score.sort_values().index.tolist()

names = ["High-Value", "Loyal", "Promising", "At-Risk", "Hibernating", "Others"]
mapping = {cluster: names[i] if i < len(names) else f"Segment {i+1}"
           for i, cluster in enumerate(order)}

rfm["SegmentName"] = rfm["Cluster"].map(mapping)
rfm.head()
```
